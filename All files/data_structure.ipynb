```python
metals=['silver','copper','gold','aluninium','gold','silver','gold','iron','gold','silver']
#Create a new list using above list which will contain same itmes in same order expect copper 
new_list =[]
for i in metals:
    if i != 'copper':
        new_list.append(i)
print(new_list)
```

```python
metals=['silver','copper','gold','aluninium','gold','silver','gold','iron','gold','silver']
new_metals = [i for i in metals if i!='copper']    #List comprehension 
print(new_metals)
```
